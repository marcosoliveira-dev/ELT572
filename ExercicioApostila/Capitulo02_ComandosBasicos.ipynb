Exercicio 01

```python
# Programa: Impressão de números de 1 a 20
# Exibir os números de 1 a 20 primeiro um abaixo do outro
# e depois um ao lado do outro.

# Parte 1 - Um abaixo do outro
print("Números de 1 a 20 (um abaixo do outro):")
for i in range(1, 21):
    print(i)

# Linha em branco para separar as duas partes
print("\nNúmeros de 1 a 20 (um ao lado do outro):")

# Parte 2 - Um ao lado do outro
for i in range(1, 21):
    print(i, end=" ")

print()  # quebra de linha final
```

Exercicio 02


```python
# Programa: Vetor de 10 números reais em ordem inversa
# Ler 10 números reais e exibi-los na ordem inversa da leitura

# Criação de uma lista vazia
numeros = []

# Leitura dos 10 números reais
for i in range(10):
    valor = float(input(f"Digite o {i+1}º número: "))
    numeros.append(valor)

# Exibição dos números na ordem inversa
print("\nNúmeros na ordem inversa:")
for i in range(9, -1, -1):
    print(numeros[i])
```

Exercicio 03


```python
# Programa: Contar consoantes em uma lista
# Ler 10 caracteres e identificar quantas e quais são consoantes

# Criação da lista vazia
caracteres = []

# Leitura dos 10 caracteres
for i in range(10):
    letra = input(f"Digite o {i+1}º caractere: ").lower()  # converte para minúsculo
    caracteres.append(letra)

# Lista 
consoantes = []

# Verificação de consoantes
for letra in caracteres:
    if letra.isalpha() and letra not in ['a', 'e', 'i', 'o', 'u']:
        consoantes.append(letra)

# Exibição dos resultados
print(f"\nForam lidas {len(consoantes)} consoantes.")
print("Consoantes:", consoantes)
```

Exercicio 04

```python
# Programa: Contar maiores de idade
# Ler a idade de 5 pessoas e informar quantas têm 18 anos ou mais

# Inicializa o contador de maiores de idade
maiores = 0

# Loop para ler a idade de 5 pessoas
for i in range(5):
    idade = int(input(f"Digite a idade da {i+1}ª pessoa: "))
    
    # Verifica se a pessoa atingiu a maior idade (18 anos ou mais)
    if idade >= 18:
        maiores += 1

# Exibe o resultado
print(f"\n{maiores} pessoa(s) atingiram a maior idade.")
```
