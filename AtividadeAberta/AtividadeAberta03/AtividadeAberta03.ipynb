## Exercicio 01


Qual o valor da variável ‘resultado’ após a execução do código? Explique sua resposta.

```python
#Exercicio 1

# Cria uma tupla contendo quatro números inteiros.
# Tuplas são estruturas de dados imutáveis em Python, ou seja, 
# depois de criadas, seus elementos não podem ser alterados.

tupla = (3,5,7,9)

# A variável 'resultado' recebe a soma de dois elementos da tupla:
# - tupla[1] acessa o elemento de índice 1 (o segundo elemento, que é 5);
# - tupla[-1] acessa o último elemento da tupla (o valor 9);
# Portanto, resultado = 5 + 9 = 14.
resultado = tupla[1] + tupla[-1]
print(resultado)
```

## Exercicio 03

Qual será o estado final do dicionário após a execução do código? Explique como os
valores foram alterados ou adicionados.

```python
# Exercicio 3

# Cria um dicionário com três pares chave-valor iniciais:
# 'a' -> 1, 'b' -> 2, 'c' -> 3.
dicionario = {'a':1, 'b':2, 'c':3}

# Aqui o valor da chave 'b' é alterado.
# O novo valor de 'b' será igual ao valor associado à chave 'c' (que é 3) somado a 2.
# Assim, dicionario['b'] passa de 2 para 3 + 2 = 5.
dicionario['b'] = dicionario['c'] + 2

# Uma nova chave chamada 'd' é adicionada ao dicionário com o valor 4.
# Dicionários permitem a inserção dinâmica de novos pares chave-valor.
dicionario['d'] = 4

# O estado final do dicionário será:
# {'a': 1, 'b': 5, 'c': 3, 'd': 4}
```

```python
print(dicionario)
```

## Exercicio 04

Qual será o conteúdo da lista resultado após a execução do loop, e qual será o
valor da variável resultado_final? Explique como o código funciona.

```python
# Exercicio 4

# Cria uma lista chamada 'numeros' contendo seis elementos inteiros.
numeros = [1, 2, 3, 4, 5, 6]

# Cria uma lista vazia chamada 'resultado', que será preenchida dentro do loop.
resultado = []

# O loop percorre os índices da lista 'numeros' de 0 até 5 (len(numeros) = 6).
for i in range(len(numeros)):
    
    # A condição 'if i % 2 == 0' verifica se o índice é par.
    # Se for par, o elemento correspondente é elevado ao quadrado (** 2)
    # e adicionado à lista 'resultado'.
    if i % 2 == 0:
        resultado.append(numeros[i] ** 2)
    else:
        # Caso contrário (índice ímpar), o elemento é multiplicado por 2
        # e o resultado é adicionado à lista.
        resultado.append(numeros[i] * 2)

# Ao final do loop, a lista 'resultado' conterá:
# Índice 0 → 1² = 1
# Índice 1 → 2×2 = 4
# Índice 2 → 3² = 9
# Índice 3 → 4×2 = 8
# Índice 4 → 5² = 25
# Índice 5 → 6×2 = 12
# Portanto, resultado = [1, 4, 9, 8, 25, 12]
print(resultado)

# A função sum(resultado) soma todos os valores da lista.
# 1 + 4 + 9 + 8 + 25 + 12 = 59
resultado_final = sum(resultado)
print(resultado_final)  # Saída: 59
```
